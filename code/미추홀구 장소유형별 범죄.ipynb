```python
import pandas as pd

# Set pandas display options to show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)


# Replace 'your_file.csv' with the actual path to your CSV file
try:
    df = pd.read_csv('/content/경찰청 인천광역시경찰청_범죄발생장소별통계 미추홀구 전처리.csv', encoding='cp949')
    print("CSV file loaded successfully!")
    display(df.head()) # Display the first few rows to confirm loading
except FileNotFoundError:
    print("Error: The file 'your_file.csv' was not found. Please provide the correct path to your CSV file.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Find the row for '미추홀서'
michuhol_row = df[df['경찰서'] == '미추홀서'].iloc[0]

# Exclude '경찰서' and '총합' columns and sort the values in descending order
sorted_values = michuhol_row.drop(['경찰서', '총합']).sort_values(ascending=False)

# Select the top 4 values and their corresponding column names
top_4_values = sorted_values.head(4)

# Display the top 4 values and their column names
print("Top 4 values for 미추홀서:")
display(top_4_values)
```

```python
# Calculate the percentage of '유흥접객업소,식당' relative to '총합' for all rows
df['유흥접객업소,식당_총합_Percentage'] = (df['유흥접객업소,식당'] / df['총합']) * 100

# Sort the DataFrame by the new percentage column in descending order
df_sorted = df.sort_values(by='유흥접객업소,식당_총합_Percentage', ascending=False)

# Display the sorted DataFrame with the relevant columns
print("DataFrame with '유흥접객업소,식당' percentage sorted in descending order:")
display(df_sorted[['경찰서', '유흥접객업소,식당', '총합', '유흥접객업소,식당_총합_Percentage']])
```
