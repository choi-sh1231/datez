# 4.** 최종 통합 코드**

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import matplotlib.font_manager as fm
import numpy as np

# 1. 범죄 데이터 로드 및 통합
df = pd.read_csv('/content/경찰청_범죄_인천_범죄유형별 분류.csv', encoding='cp949')

# '범죄대분류' 기준으로 그룹화 및 합계 계산
grouped_df = df.groupby('범죄대분류').agg({
    '범죄중분류': 'first', # 또는 다른 적절한 방법으로 대표값 선택
    '인천중구': 'sum',
    '인천동구': 'sum',
    '인천미추홀구': 'sum',
    '인천연수구': 'sum',
    '인천남동구': 'sum',
    '인천부평구': 'sum',
    '인천계양구': 'sum',
    '인천서구': 'sum',
    '인천강화군': 'sum',
    '인천옹진군': 'sum'
}).reset_index()

# 2. CCTV 데이터 로드 및 전처리
df_cctv = pd.read_csv('/content/인천 군구별 CCTV 수.csv', encoding='cp949', header=1)
df_cctv.columns = ['군구', 'CCTV 수']
df_cctv = df_cctv.drop(0).reset_index(drop=True) # 인덱스 재설정 추가

# 폰트 설정 (한글 깨짐 방지)
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf

fm.fontManager.addfont('/usr/share/fonts/truetype/nanum/NanumGothicBold.ttf')
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호 깨짐 방지

# 3. 데이터 통합 (범죄 및 CCTV)
column_mapping = {
    '인천중구': '중구',
    '인천동구': '동구',
    '인천미추홀구': '미추홀구',
    '인천연수구': '연수구',
    '인천남동구': '남동구',
    '인천부평구': '부평구',
    '인천계양구': '계양구',
    '인천서구': '서구',
    '인천강화군': '강화군',
    '인천옹진군': '옹진군'
}
grouped_df = grouped_df.rename(columns=column_mapping)

melted_df = grouped_df.melt(
    id_vars=['범죄대분류', '범죄중분류'],
    var_name='군구',
    value_name='범죄수'
)

merged_df = pd.merge(melted_df, df_cctv, on='군구', how='inner')

# 4. 상관관계 분석 (범죄 대분류별)
correlation_by_crime_type = merged_df.groupby('범죄대분류').apply(lambda x: x['범죄수'].corr(x['CCTV 수'])).reset_index(name='CCTV 상관관계')

# 5. 상관관계 결과 시각화 (범죄 대분류별)
plt.figure(figsize=(12, 7))
plt.bar(correlation_by_crime_type['범죄대분류'], correlation_by_crime_type['CCTV 상관관계'])
plt.title('범죄 대분류별 CCTV 상관관계')
plt.xlabel('범죄 대분류')
plt.ylabel('CCTV 상관관계')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 6. 음의 상관관계를 띄는 대분류 제거 및 데이터 통합
negative_correlation_crime_types = correlation_by_crime_type[correlation_by_crime_type['CCTV 상관관계'] < 0]['범죄대분류'].tolist()
filtered_merged_df = merged_df[~merged_df['범죄대분류'].isin(negative_correlation_crime_types)]
total_crime_by_district = filtered_merged_df.groupby('군구')['범죄수'].sum().reset_index(name='총 범죄수')

# 7. 통합된 범죄 수와 CCTV 수 간의 상관관계 분석 및 시각화
merged_total_crime_cctv = pd.merge(total_crime_by_district, df_cctv, on='군구', how='inner')
correlation_total_crime_cctv = merged_total_crime_cctv['총 범죄수'].corr(merged_total_crime_cctv['CCTV 수'])

plt.figure(figsize=(10, 6))
plt.scatter(merged_total_crime_cctv['CCTV 수'], merged_total_crime_cctv['총 범죄수'])

# 회귀선 추가
z = np.polyfit(merged_total_crime_cctv['CCTV 수'], merged_total_crime_cctv['총 범죄수'], 1)
p = np.poly1d(z)
plt.plot(merged_total_crime_cctv['CCTV 수'], p(merged_total_crime_cctv['CCTV 수']), color='red', linestyle='--')


plt.title('인천 군구별 CCTV 수와 (음의 상관관계 항목 제외) 총 범죄 발생 수 간의 관계')
plt.xlabel('CCTV 수')
plt.ylabel('(음의 상관관계 항목 제외) 총 범죄 발생 수')
plt.grid(True)
plt.show()

display("음의 상관관계 항목 제외 후 총 범죄수와 CCTV 수 간의 상관관계:", correlation_total_crime_cctv)
```
